## What is ABZU?

ABZU is a **sovereign development environment** running on the BEAM (Erlang VM) that embeds AI reasoning, cryptographic audit trails, and a trust-enforced package registry directly into the IDE runtime.

### Architecture

| Layer | Technology | Role |
|---|---|---|
| **IDE Shell** | Phoenix LiveView | Real-time browser IDE, zero JS framework |
| **Editor** | Monaco (VS Code engine) | Full Elixir syntax, Ctrl+Enter run |
| **Runner** | `Code.eval_string/2` on BEAM | Live Elixir execution in OTP sandbox |
| **AI** | BOB Agent → IBM Gamma vLLM | Complete / Explain / Repair — sovereign model |
| **Audit** | WORM Chain (DETS + djb2 seal) | Append-only, every action sealed |
| **Registry** | Sovereign NPM (snapkittywest org) | WORM-seals every install, flags foreign packages |

### The WORM Seal Format

Every action in ABZU produces an immutable seal:
```
{djb2_hash}-worm-{payload_len_hex}
```
The chain is append-only — no delete, no overwrite. Stone 3 signed: `29b86e67-worm-162`

### Why BEAM?

OTP gives you fault-tolerant AI supervision for free. BOB is a GenServer — if it crashes, the supervisor restarts it in milliseconds. The IDE never goes down even when BOB is reasoning about broken code.

````python
# ══════════════════════════════════════════════════════════════════
# WORM CHAIN DEMO — run this to see the ledger in action
# Each line is cryptographically chained to the previous
# ══════════════════════════════════════════════════════════════════

import time, json

def djb2(s):
    h = 5381
    for c in s:
        h = ((h << 5) + h + ord(c)) & 0xFFFFFFFF
    return format(h, '08x')

class WormChain:
    def __init__(self):
        self.entries = []
        self.prev = '0' * 16

    def seal(self, action, payload):
        ts = int(time.time() * 1000)
        body = f"{self.prev}:{action}:{json.dumps(payload)}:{ts}"
        h = djb2(body)
        seal_id = f"{h}-worm-{hex(len(self.entries))[2:]}"
        entry = {'seal': seal_id, 'action': action, 'prev': self.prev, 'ts': ts}
        self.entries.append(entry)
        self.prev = h
        return entry

    def verify(self):
        prev = '0' * 16
        for i, e in enumerate(self.entries):
            if e['prev'] != prev:
                return False, f"Chain broken at entry #{i} — seal {e['seal']}"
            prev = e['seal'].split('-worm-')[0]
        return True, f"Chain intact — {len(self.entries)} entries verified"

# ── BUILD THE CHAIN ───────────────────────────────────────────────
chain = WormChain()

events = [
    ('ide_start',     {'product': 'ABZU', 'version': '1.0.0'}),
    ('beam_run',      {'code': 'IO.puts("Hello, sovereign")', 'result': 'Hello, sovereign'}),
    ('bob_complete',  {'prompt_tokens': 142, 'model': 'ibm-gamma'}),
    ('pkg_install',   {'package': 'snapkitty-mcp', 'status': 'sovereign'}),
    ('pkg_install',   {'package': 'lodash', 'status': 'flagged — external'}),
    ('beam_run',      {'code': 'Enum.map(1..5, &(&1 * &1))', 'result': '[1, 4, 9, 16, 25]'}),
    ('bob_repair',    {'error': 'UndefinedFunctionError', 'fixed': True}),
    ('session_end',   {'duration_ms': 342000}),
]

print('ABZU WORM CHAIN · LIVE SESSION DEMO')
print('═' * 60)
for action, payload in events:
    e = chain.seal(action, payload)
    print(f"  {e['seal']}")
    print(f"  {action:>20} ← {list(payload.keys())}")
    print()

ok, msg = chain.verify()
print(f"{'✓' if ok else '✗'} {msg}")

# ── MUTATION ATTACK — try to rewrite history ──────────────────────
print()
print('── MUTATION ATTACK ──────────────────────────────────────────')
print('Tampering with entry #2: changing bob_complete model to "backdoor"...')

import copy
tampered = copy.deepcopy(chain)
tampered.entries[2]['action'] = 'bob_complete_TAMPERED'

ok2, msg2 = tampered.verify()
print(f"{'✓' if ok2 else '✗'} {msg2}")
print()
print('The djb2 chain seal catches any mutation.')
print('No delete, no overwrite, no rewrite of history.')
````

### The Problem

Every AI IDE today is a wrapper around VS Code + an API call. The AI is external — it has no visibility into the runtime, no persistence across sessions, and no accountability for what it suggests.

### The Solution — ABZU

ABZU embeds AI **inside** the runtime:

- **BOB runs on OTP** — supervised, fault-tolerant, co-resident with the code it's reasoning about
- **WORM chain** — every edit, every run, every BOB response is cryptographically sealed. The ledger cannot be altered retroactively
- **Sovereign registry** — packages outside the trust mesh are flagged before install, not after
- **IBM Gamma vLLM** — your own model weights, your own server, zero third-party API dependency
